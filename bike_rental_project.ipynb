By providing a stable supply of rental bikes, the system can enhance mobility comfort for the public and reduce waiting time, leading to greater customer satisfaction and accurately predicting bike demand can help bike sharing companies optimize operations including bike availability, pricing, strategies, and marketing efforts by considering demand Based on various external factors such as weather, season, holiday etc..,

### 3.Feature Enineering

### 4.Model Development

#### Splitting Data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
target = df_scaled['cnt']
feature = df_scaled.drop('cnt', axis=1)
```

```python
target
```

```python
feature
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature, target, train_size=0.75, random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Building Model

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
DT = DecisionTreeRegressor(splitter='random',max_depth=3)
```

```python
DT.fit(x_train,y_train)
```

```python
yp_DT= DT.predict(x_test)
```

```python
r2_score(y_test, yp_DT)
```

```python
plt.figure(figsize=(15,9))
plot_tree(DT, filled=True)
plt.show()
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF = RandomForestRegressor(n_estimators=100)
```

```python
RF.fit(x_train,y_train)
```

```python
yp_RF = RF.predict(x_test)
```

```python
r2_score(y_test, yp_RF)
```

### 6.Model Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae_DT = mean_absolute_error(y_test, y_predDTgs)
mse_DT = mean_squared_error(y_test,y_predDTgs )
rmse_DT = np.sqrt(mse_DT)

print(' MAE for DecisionTree:', round(mae_DT, 3))
print(' MSE for DecisionTree:', round(mse_DT, 3))
print('RMSE for DecisionTree:', round(rmse_DT, 3))
```

### Random Forest

```python
mae_RF = mean_absolute_error(y_test, y_predRFgs)
mse_RF = mean_squared_error(y_test, y_predRFgs)
rmse_RF = np.sqrt(mse_RF)

print(' MAE for RandomForest:', round(mae_RF, 3))
print(' MSE for RandomForest:', round(mse_RF, 3))
print('RMSE for RandomForest:', round(rmse_RF, 3))
```

### GradientBoosting

```python
mae_GB = mean_absolute_error(y_test, y_predGBgs)
mse_GB = mean_squared_error(y_test, y_predGBgs)
rmse_GB = np.sqrt(mse_GB)

print(' MAE for GradientBoosting:', round(mae_GB, 3))
print(' MSE for GradientBoosting:', round(mse_GB, 3))
print('RMSE for GradientBoosting:', round(rmse_GB, 3))
```
